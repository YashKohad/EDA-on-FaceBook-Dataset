# **6. Data Pre-Processing**
---


- This section is emphasised on performing data manipulation over unstructured data for further processing and analysis.

- To modify unstructured data to structured data you need to verify and manipulate the integrity of the data by:
  - Handling missing data,

  - Handling redundant data,

  - Handling inconsistent data,

  - Handling outliers,

  - Handling typos

```python
# Code to set an option to display all rows in any column while displaying results post cell execution
pd.set_option('display.max_rows', None)

# Check no. of missing data rows in tenure column
df['tenure'].isnull().sum()
```

```python
# Finding median of tenure
df.tenure.median()
```

```python
#Replace missing values with median of tenure column
df['tenure'] = df['tenure'].fillna(df.tenure.median())
```

```python
# Confirm that now there are only no missing data in tenure column
df['tenure'].isnull().sum()
```

```python
# Converting tenure from float to int
df['tenure']=df['tenure'].astype('int64')
```

```python
# Checking id tenure has been converted to int from float
df.head()
```

```python
# In 'gender' column, we have 175 missing data(<0.1%) , it's better to drop these rows with missing data as it is difficult to get this data from actual source
# Creating a new data frame called 'new_df' and storing the results post dropping of rows with missing data
# Here axis=0, means row wise
new_df = df.dropna(axis = 0, how ='any') 
```

```python
new_df.shape
# 99003-175
```

```python
# Searching the rows with tenure as 0 
new_df[new_df['tenure']==0]
# 70 rows
```

```python
# Users with tenure as zero seems unlikely, so removing these rows
temp_df=new_df.drop(new_df[new_df['tenure']==0].index)
new_df = temp_df

# Removed rows with tenure as zero
```

```python
# 98828-70
new_df.shape
```

```python
# Florence Detlor, who at 101 years old was the oldest registered Facebook user (died in 2016 with 105 age)
# So we consider users with age >105 as invalid data and remove such rows from our dataset.
new_df[new_df.age>105].count()
# There are 2115 no. of people with age>105
```

```python
temp_df=new_df.drop(new_df[new_df.age>105].index)
new_df=temp_df
# Data of users with age>105 removed from our dataset.
```

```python
# 98758-2115
new_df.shape
```

```python
# Dividing the age of users into a group of 10 and storing it in a new column
age_groups=['10-20','21-30','31-40','41-50','51-60','61-70','71-80','81-90','91-100','>100']
new_df['age_group'] = pd.cut(new_df.age,bins=np.arange(10,111,10),labels=age_groups,right=True)
```

```python
# Confirming the addition of the column in our dataset
new_df.head()
```
